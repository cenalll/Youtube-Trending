```python
import json

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
%matplotlib inline
```

```python
cate_id_to_category={}

with open('US_category_id.json', 'r') as f:
    data=json.load(f)
    for category in data['items']:
        cate_id_to_category[category['id']] = category['snippet']['title']
```

```python
cate_id_to_category
```

```python
youtube_us.insert(4,'category', youtube_us['category_id'].astype(str).map(cate_id_to_category))
youtube_us[['category_id', 'category']].head(10)
```

```python
youtube_us.sort_values('views',ascending=False,inplace=True)
x_axis=youtube_us.head(10)
x_axis
```

```python
youtube_us = youtube_us[~youtube_us['title'].duplicated(keep='last')] 
youtube_us['title'].duplicated().any()
```

```python
def max_visualized(youtube_us,column,num=10):
    max_youtube_us=youtube_us.sort_values(column,ascending=False).iloc[:num]
    
    fig,ax=plt.subplots()
    ax.barh(max_youtube_us['title'],max_youtube_us[column])
    
    ax.set(xlabel=column,title ='Top %s %s of video' %(num,column))
```

```python
max_visualized(youtube_us,'views',20)
```

```python
max_visualized(youtube_us,'likes',20)    
```

```python
max_visualized(youtube_us,'dislikes',20)
```

```python
max_visualized(youtube_us,'comment_count',20)
```

```python
sns.factorplot('category',data=youtube_us,size = 7.5, aspect = 3,hue='category',kind='count',legend_out=True)
```

```python
youtube_us['views_log'] = np.log(youtube_us['views'] + 1)

plt.figure(figsize = (20,8))
plt.title('Views distribution by category')
plt.xticks(rotation=20)
sns.boxplot(x='category',y='views_log',data=youtube_us, width=0.5,palette='Set1')
```

```python
youtube_us['likes_log'] = np.log(youtube_us['likes'] + 1)

plt.figure(figsize = (20,8))
plt.title('Likes distribution by category')
plt.xticks(rotation=20)
sns.boxplot(x='category',y='likes_log',data=youtube_us, width=0.5,palette='Set1')
```

```python
youtube_us['dislikes_log'] = np.log(youtube_us['dislikes'] + 1)

plt.figure(figsize = (20,8))
plt.title('Dislikes distribution by category')
plt.xticks(rotation=20)
sns.boxplot(x='category',y='dislikes_log',data=youtube_us, width=0.5,palette='Set1')
```

```python
sns.lmplot(x='likes',y='dislikes',data=youtube_us,size=8, scatter_kws={'marker':'o','color':'pink'},line_kws={'linewidth':1,'color':'black'})
```

```python
youtube_us['comment_count_log'] = np.log(youtube_us['comment_count'] + 1)

plt.figure(figsize = (20,8))
plt.title('Comments distribution by category')
plt.xticks(rotation=20)
sns.boxplot(x='category',y='comment_count_log',data=youtube_us, width=0.5,palette='Set1')
```

```python
youtube_us['like_rate']=youtube_us['likes']/youtube_us['views']*100
youtube_us['dislike_rate']=youtube_us['dislikes']/youtube_us['views']*100
youtube_us['comment_rate']=youtube_us['comment_count']/youtube_us['views']*100
```

```python
plt.figure(figsize = (10,8))
plt.title('Rate distribution')
sns.distplot(youtube_us['like_rate'],color='red',hist=False,label='Likes')
sns.distplot(youtube_us['dislike_rate'],color='green',hist=False,label='Dislikes')
sns.distplot(youtube_us['comment_rate'],color='blue',hist=False,label='Comments')
```

```python
youtube_us['trending_date'] = pd.to_datetime(youtube_us['trending_date'], format='%y.%d.%m')
```

```python
youtube_us['publish_time'] = pd.to_datetime(youtube_us['publish_time'], format='%Y-%m-%dT%H:%M:%S.%fZ',errors='coerce')
```

```python
youtube_us.insert(4, 'publish_date', youtube_us['publish_time'].dt.date)
youtube_us['publish_time'] = youtube_us['publish_time'].dt.time
youtube_us[['publish_date', 'publish_time']].head()
```

```python
youtube_us['publish_day']=youtube_us['publish_date'].apply(lambda time : time.day)
```

```python
publish_date_vs_category=youtube_us.groupby(['category','publish_day'])['views'].sum()
pub_vs_cate=publish_date_vs_category.unstack().fillna(0)

plt.figure(figsize=(12,12))
plt.title('Category VS Public date')
sns.heatmap(pub_vs_cate)
```

```python
youtube_us['publish_month']=youtube_us['publish_date'].apply(lambda time : time.month)
```
